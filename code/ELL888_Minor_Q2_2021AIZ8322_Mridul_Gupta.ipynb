Since the dataset is really big ($60000$) images, I'll randomly choose $100$ of these and work with them. Also, since in our problem we only had two classes ($y\in\{0,1\}$), I'll be going choosing images only of class $0$ and $1$.

Also note, the images are of size $28\times 28$ which will be converted into a $1\text{D}$ vector $\in\mathbb{R}^{784}$. So, for our data matrix $X\in \mathbb{R}^{n\times d}$, $n$ will be $100$ and $d$ will be $784$.

Since $d$ is going to be large, I also try to compress the data using PCA before filtering the classes. Chosen value of the projected size $15$. So now, data matrix is $\mathbf{X}\in\mathbb{R}^{100\times 15}$.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
images = (images.reshape(images.shape[0],-1)-127)/127
images = pca.fit_transform(images)
```

```python
n=100
```

```python
filter_ind = ((labels == 0) | (labels == 1))
images = images[filter_ind]
labels = labels[filter_ind]

assert np.all((labels==0) + (labels==1))
```

```python
chosen = np.random.permutation(labels.shape[0])[:n]
chosen_images = images[chosen].reshape(n,-1)
chosen_labels = labels[chosen]
```

And out of these, I'll randomly drop the labels for $40\%$ of the population

```python
pct_ign = 0.4
ignore = np.random.permutation(n)[:int(n*pct_ign)]
ignore_flags = np.zeros(chosen_labels.shape,dtype=bool)
ignore_flags[ignore] = True
```

#### Distance metric
First, we begin with the Euclidean distance
$$D(x^i,x^j) = \biggl(\sum_{k=1}^d(x^i_k-x^j_k)^2\biggr)^{\frac{1}{2}}$$
and modify this distance when we have both labels $y^i$ corresponding to $x^i$ and $y^j$ associated with $x^j$ as follows:
$$\tilde{D}(x^i,x^j) = D(x^i,x^j)+k(1\{y^i=1-y^j\}-1\{y^i=y^j\})$$


```python
k = 6
def distance(x1,x2,y1=None,y2=None):
    assert type(x1) == np.ndarray
    assert type(x2) == np.ndarray
    global k
    def eucl(x1,x2):
        return np.sqrt(np.sum(np.square(x1-x2)))
    d = eucl(x1, x2)
    if y1 is not None and y2 is not None:
#         print(d,y1,y2)
        d += k * (-1 if y1==y2 else 1)
        d = max(0,d)
#         print(d)
    return d
```

```python
idx1 = np.random.randint(1,chosen_labels.shape[0])
idx2 = np.random.randint(1,chosen_labels.shape[0])
print(chosen_labels[idx1],chosen_labels[idx2])
if ignore_flags[idx1] or ignore_flags[idx2]:
    print(distance(chosen_images[idx1],chosen_images[idx2]))
else:
    print(distance(chosen_images[idx1],chosen_images[idx2],chosen_labels[idx1],chosen_labels[idx2]))
```

#### Constructing the second order matrix

```python
def make_sec_matrix():
    global chosen_images, chosen_labels, ignore_flags
    mat = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            if ignore_flags[i] or ignore_flags[j]:
                mat[i,j] = distance(chosen_images[i],chosen_images[j])
            else:
                mat[i,j] = distance(chosen_images[i],chosen_images[j],chosen_labels[i],chosen_labels[j])
    return mat
```

```python
sec_ord_matrix = make_sec_matrix()
```

### Graph construction

Finally, using the second order matrix, we can learn graph. The method I use is taken from [Diffusion Convolutional Recurrent Neural Networks](https://arxiv.org/abs/1707.01926) [1]. They use the Gaussian Kernel method along with a thresholding as:
$$\large
W_{ij} = \left\{
\begin{array}\\
\exp\left(-\frac{d(x^i,x^j)^2}{\sigma^2}\right) & \mbox{if } \ d(x^i,x^j)\le \kappa \\
0 & \mbox{otherwise.}
\end{array}
\right.
$$
where $\kappa$ is the threshold and $\sigma$ is the standard deviation of the distances.

[1] Li, Yaguang, Rose Yu, Cyrus Shahabi, and Yan Liu. "Diffusion convolutional recurrent neural network: Data-driven traffic forecasting." arXiv preprint arXiv:1707.01926 (2017).

```python
non_zero = sec_ord_matrix[sec_ord_matrix!=0]
sig = np.std(non_zero.reshape(-1))
mu = np.mean(non_zero.reshape(-1))
print(mu,sig)
threshold = mu - (sig * 1.2)
```

```python
W = np.exp(-np.square(sec_ord_matrix)/(sig**2))
W[sec_ord_matrix>threshold]=0
```

#### Visualizing the graph
Now, let's visualize the graph created using networkx

```python
import networkx as nx
```

```python
g = nx.Graph()
for i in range(n):
    for j in range(n):
        if W[i,j] != 0:
            g.add_edge(i,j,weight=W[i,j])
```

```python
fig = plt.figure(figsize=(10,10))
# pos = nx.kamada_kawai_layout(g)
pos = nx.shell_layout(g)
options = {"edgecolors": "tab:gray", "node_size": 500, "alpha": 0.9}
nx.draw_networkx_nodes(g, pos, nodelist=[_ for _ in range(n)], node_color="tab:red", **options)
ec=nx.draw_networkx_edges(g, pos, width=1.0, alpha=0.5)
self_loops = plt.gca().patches
for l in self_loops:
    l.set_visible(False)
seen_labels = {}
for i in range(n):
    if not ignore_flags[i]:
        seen_labels[i]=chosen_labels[i]
nx.draw_networkx_labels(g, pos, seen_labels, font_size=22, font_color="whitesmoke")
hidden_labels = {}
for i in range(n):
    if ignore_flags[i]:
        hidden_labels[i]=chosen_labels[i]
nx.draw_networkx_labels(g, pos, hidden_labels, font_size=22, font_color="black")
plt.tight_layout()
plt.axis("off")
plt.savefig("Generated_graph_{}.png".format(n))
print("")
```
