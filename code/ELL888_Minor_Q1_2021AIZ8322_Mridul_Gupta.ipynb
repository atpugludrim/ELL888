````python
import copy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
np.set_printoptions(suppress=True)
````

````python
data = read_seeds("SEEDS/seeds_dataset.txt")
````

````python
print(data.shape)
````

This data is complete, nothing missing. So I'll create a synthetic data with some attributes missing at random.

````python
mask = np.random.random(data.shape)
mask = mask < 0.01 # 1 % missing at random (iid across samples and features)
````

````python
print(np.sum(mask))
print(np.any(mask))
mask_backup = copy.deepcopy(mask)
````

In numpy, the standard way to have arrays with missing data is to use a masked array, and the valid entries can be accessed using

```data[~data.mask]```

````python
masked_data = np.ma.array(copy.deepcopy(data),mask=mask)
````

````python
print(np.mean(data,0))
print(np.mean(masked_data,0))
print(np.std(data,0))
print(np.std(masked_data,0))
````

### Using Gaussian Mixture Models to learn the density of this data

````python
from sklearn.mixture import GaussianMixture
````

We need one for each feature $k\in \{1,\dotsc,d\}$ $\text{(}\mathbf{X}\in\mathbb{R}^{n\times d}\text{)}$.

````python
nc = 5
````

````python
gmms = [GaussianMixture(n_components=nc,max_iter=1000,covariance_type='full') for _ in range(data.shape[1])]
````

````python
for k in range(data.shape[1]):
    features_slice = masked_data[:,k]
    valid_feature_values = features_slice[~features_slice.mask].data
    gmms[k].fit(valid_feature_values.reshape(-1,1))
````

````python
# for k in range(data.shape[1]):
#     print("For feature",k)
#     print(gmms[k].means_)
#     print(gmms[k].covariances_)
````

For sampling from these distributions now

````python
import scipy.stats as ss
````

````python
for k in range(data.shape[1]):
    mask = masked_data.mask[:,k]
    if np.any(mask):
        idx = np.nonzero(mask)
        for i in idx[0]:
            mixture_idx = np.random.choice(nc,size=1,replace=True,p=np.ones(nc)/nc).item()
            generated_data = ss.norm.rvs(gmms[k].means_[mixture_idx],gmms[k].covariances_[mixture_idx])
            masked_data[i,k] = float("{:.3f}".format(generated_data))
````

````python
np.any(masked_data.mask) # every data point is now filled
````

````python
imputed_data = masked_data.data
````

````python
np.sqrt(np.sum(np.square(imputed_data-data)))
````

````python
data
````

````python
imputed_data
````

````python
data[mask_backup]
````

````python
imputed_data[mask_backup]
````

### Graph construction

````python
from scipy.spatial.distance import pdist, squareform
````

````python
sec_ord_matrix = squareform(pdist(imputed_data))
````

````python
non_zero = sec_ord_matrix[sec_ord_matrix!=0]
sig = np.std(non_zero.reshape(-1))
mu = np.mean(non_zero.reshape(-1))
print(mu,sig)
threshold = mu - sig * 1.5
````

````python
W = np.exp(-np.square(sec_ord_matrix)/(sig**2))
W[sec_ord_matrix>threshold] = 0
````

Network visualization using networkx

````python
import networkx as nx
````

````python
G = nx.Graph()
n = W.shape[0]
for i in range(n):
    for j in range(n):
        if W[i,j] != 0:
            G.add_edge(i,j,weight=W[i,j])
````

````python
fig = plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
options = {"node_size":60,"alpha":0.7}
nx.draw_networkx_nodes(G,pos,node_color="tab:cyan",**options)
ec = nx.draw_networkx_edges(G,pos,alpha=0.2)
self_loops = plt.gca().patches
for l in self_loops:
    l.set_visible(False)
plt.tight_layout()
plt.savefig("Seeds.png")
plt.axis("off")
````
