[Dataset](https://archive.ics.uci.edu/ml/datasets/Hepatitis) being used has 14 categorical attributes, 4 integer attributes and 2 real attributes.

```python
from data.read_hepatitis import read_hepatitis
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cdf, idf, rdf = read_hepatitis('HEPATITIS/hepatitis.data')
```

```python
cdf_one_hot_encoded = pd.get_dummies(cdf)
```

```python
W_i, W_c, W_r = 1, 1, 1
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
categ = cdf_one_hot_encoded.values
dist_c = squareform(pdist(categ,"hamming"))*categ.shape[1]
dist_i = squareform(pdist(idf.values,"euclidean"))
dist_r = squareform(pdist(rdf.values,"euclidean"))
```

```python
dist = (W_i*dist_i+W_c*dist_c+W_r*dist_r)/(W_i+W_c+W_r)
```

```python
nz = dist[dist!=0]
sig = np.std(nz.reshape(-1))
mu = np.mean(nz.reshape(-1))
print(mu,sig)
threshold = mu - sig * 0.87
```

```python
W = np.exp(-np.square(dist)/(sig**2))
W[dist>threshold]=0
```

```python
W
```

```python
np.sum(W==0)
```

```python
import networkx as nx
```

```python
G = nx.Graph()
n = cdf.shape[0]
for i in range(n):
    for j in range(n):
        if W[i,j] != 0:
            G.add_edge(i,j,weight=W[i,j])
```

```python
fig = plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
options = {"node_size":60,"alpha":0.7}
nx.draw_networkx_nodes(G,pos,node_color="tab:green",**options)
ec = nx.draw_networkx_edges(G,pos,alpha=0.2)
self_loops = plt.gca().patches
for l in self_loops:
    l.set_visible(False)
plt.tight_layout()
plt.savefig("hepatitis.png")
plt.axis("off")
```
